```python
import pickle
import sys
from simulation import Simulation
from os import listdir
from os.path import isfile, join
import matplotlib.pyplot as plt
```

```python
def get_task_results(mypath):
    onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
    if '.DS_Store' in onlyfiles:
        onlyfiles.remove('.DS_Store')
    sorted = onlyfiles.sort(key=lambda x:int(x.split('_')[1]))
    onlyfiles.sort(key=lambda x:int(x.split('_')[1]))
    results = []
    for i in onlyfiles:
        results.append(pickle.load(open(join(mypath,i),'rb')))
    return results
```

```python
rtrl_results = get_task_results(mypath ='../cluster_results/rtrl_test/')
```

```python
fig, axs = plt.subplots(2, 2,figsize=(20,10))

for r in rtrl_results:
    avg_loss = r['average_loss']

    if r['config']['lr'] == 0.1:
        legend = 'units{}_lr{}'.format(r['config']['units'],r['config']['lr'])
        axs[0, 0].plot(avg_loss, label =legend)
    axs[0, 0].legend()
    axs[0, 0].set_ylim(0.4,0.8)
    axs[0, 0].set_title('RTRL LSTM Average Loss')
    
    if r['config']['lr'] == 0.01:
        legend = 'units{}_lr{}'.format(r['config']['units'],r['config']['lr'])
        axs[0, 1].plot(avg_loss,label = legend)
    axs[0, 1].legend()
    axs[0, 1].set_ylim(0.4,0.8)
    axs[0, 1].set_title('RTRL LSTM Average Loss')

    if r['config']['lr'] == 0.001:
        legend = 'units{}_lr{}'.format(r['config']['units'],r['config']['lr'])
        axs[1, 0].plot(avg_loss,label=legend)        
    axs[1, 0].legend()
    axs[1, 0].set_ylim(0.4,0.8)
    axs[1, 0].set_title('RTRL LSTM Average Loss')

    if r['config']['lr'] == 0.0001:
        legend = 'units{}_lr{}'.format(r['config']['units'],r['config']['lr'])
        axs[1, 1].plot(avg_loss,label =legend)       
    axs[1, 1].legend()
    axs[1, 1].set_ylim(0.4,0.8)
    axs[1, 1].set_title('RTRL LSTM Average Loss')

    print(min(avg_loss),legend)
```
